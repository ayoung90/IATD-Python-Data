```python
import pandas
import matplotlib.pyplot as plt
import numpy as np

# Plotting cumulative rainfall in 2020

wentworth = pandas.read_csv('data/rainfall/IDCJAC0009_047045_1800_Data.csv')

mask_2020 = wentworth["Year"] == 2020
wentworth_2020 = wentworth[mask_2020]

wentworth_2020['cumulative rainfall'] = wentworth_2020['Rainfall amount (millimetres)'].cumsum()
wentworth_2020 = wentworth_2020.reset_index(drop=True)

wentworth_2020['cumulative rainfall'].plot(legend=True, title='Rainfall in wentworth 2020')

plt.xlabel('Day in year')
plt.ylabel('Cumulative rainfall')

# Plotting the regression line
x = wentworth_2020.index
y = wentworth_2020['cumulative rainfall']

coefficients = np.polyfit(x, y, 1)

poly1d_fn = np.poly1d(coefficients)

wentworth_2020['predictions'] = poly1d_fn(x)

wentworth_2020['predictions'].plot(legend=True)
```
